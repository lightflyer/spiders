```python
import re

def get_highest_pay(s):
    """ 10k-20k 10以上"""
    return max([int(_) for _ in re.findall('\d+', s)])
    
def get_lowest_pay(s):
    return min([int(_) for _ in re.findall('\d+', s)])

salary_df = pd.DataFrame([df.district, df.salary]).T
salary_df['salary'] = salary_df['salary'].apply(get_highest_pay)


salary_df.head()
```

```python
salary_df['salary'].astype('int32')
salary_df = salary_df.groupby('district')['salary'].mean()
salary_df
```

```python
xs,ys = salary_df.index, salary_df

for x,y in zip(xs, ys):
    plt.text(x, y, '%.02f' % y, ha='center', va= 'bottom',fontsize=8)
    
plt.title('深圳各区最高工资平均数')
plt.bar(xs, ys)
plt.savefig('/Users/michael/Downloads/深圳各区最高工资平均数.jpg')
```

```python
salary_df = pd.DataFrame([df.district, df.salary]).T
salary_df['salary'] = salary_df['salary'].apply(get_lowest_pay)
salary_df=salary_df[~salary_df['district'].isin(['天河区', '黄浦区'])]
salary_df['salary'].astype('int32')
salary_df = salary_df.groupby('district')['salary'].mean()
xs,ys = salary_df.index, salary_df

for x,y in zip(xs, ys):
    plt.text(x, y, '%.02f' % y, ha='center', va= 'bottom',fontsize=8)
    
plt.title('深圳各区最低工资平均数')
plt.bar(xs, ys)
plt.savefig('/Users/michael/Downloads/深圳各区最低工资平均数.jpg')
```
